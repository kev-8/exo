### Incorporating news data into dataset

```python
gp_news = pd.read_csv('/Users/KevinLubin/Desktop/dosi/exo/data/gp_news.csv')
```

```python
gp_news.head()
```

```python
gp_news.shape
```

```python
#dropping duplicate news articles

gp_news = gp_news.drop_duplicates(subset=['Sourceurl'], keep='last').reset_index(drop=True)
gp_news = gp_news.rename(columns={"SQLDATE": "Date"})
gp_news['Date'] = gp_news['Date'].astype(str)
gp_news.shape
```

```python
gp_news.head()
```

```python
#creating new variables for two sentiment variables for each Event by Date

tone = gp_news.groupby(['Date','EventRootCode','Quadclass'], as_index = False)['AvgTone'].mean()
tone = tone.rename(columns={"EventRootCode":"event","Quadclass":"class","AvgTone":"tone"})
tone['event'] = tone['event'].astype(str)

goldstein = gp_news.groupby(['Date','EventRootCode','Quadclass'], as_index = False)['GoldSteinScale'].mean()
goldstein = goldstein.rename(columns={"EventRootCode":"event","Quadclass":"class","GoldSteinScale":"goldstein"})
goldstein['event'] = goldstein['event'].astype(str)\
```

```python
tone.head()
```

```python
goldstein.head()
```

```python
#using codes for events to rename event column

cameo_codes = pd.read_csv('/Users/KevinLubin/Desktop/dosi/exo/data/cameo_codes.csv',dtype={'code':str})
cameo_codes['event'] = cameo_codes['event'].str.lower()
cameo_codes = dict(zip(cameo_codes.code,cameo_codes.event))
tone['event'].replace(cameo_codes, inplace=True)
goldstein['event'].replace(cameo_codes, inplace=True)
```

```python
#organizing the data by date for each Event category 

tone = tone.pivot_table(
    values='tone', index='Date', columns='event', aggfunc='mean')

goldstein = goldstein.pivot_table(
    values='goldstein', index='Date', columns='event', aggfunc='mean')
```

```python
#resetting indexes to make indexes columns

tone.reset_index(level=tone.index.names, inplace=True)
goldstein.reset_index(level=goldstein.index.names, inplace=True)
```

```python
tone.head()
```

```python
#summary statistics

tone.describe()
```

```python
#dropping column for lack of data

tone.drop(['mass_violence'], axis=1, inplace=True)
```

```python
#summary statistics

goldstein.describe()
```

```python
#dropping columns for lack of variance

goldstein.drop(['demand','disapprove','investigate','mass_violence'], axis=1, inplace=True)
```

```python
#changing date to datetime object to align with other dataframes for merging 

tone['datetime'] = pd.to_datetime(tone['Date'])
tone = tone.set_index('datetime')
tone.drop(['Date'], axis=1, inplace=True)

goldstein['datetime'] = pd.to_datetime(goldstein['Date'])
goldstein = goldstein.set_index('datetime')
goldstein.drop(['Date'], axis=1, inplace=True)
```

```python
#adding suffix to column names so there is no confusion in merged dataframe

tone = tone.add_suffix('_tone')
goldstein = goldstein.add_suffix('_goldstein')
```

```python
df.describe()
```

```python
#create correlation matrix
corr_matrix = df.corr().abs()

#select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

#find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
to_drop

# Drop those highly correlated features 
df.drop(to_drop, axis=1, inplace=True)
```
